```python
import matplotlib.pyplot as plt
import sandbox_fm.depth
import collections
import numpy as np
%matplotlib inline
```

```python
images = sandbox_fm.depth.calibrated_depth_images()
img = next(images)
```

```python
buffer = collections.deque(maxlen=10)
for i, img in enumerate(images):
    buffer.append(img)
    if i > 100:
        break
perc = np.percentile(buffer, q=25, axis=0)
plt.imshow(perc, cmap='viridis')
```
